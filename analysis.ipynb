```python
#B
import csv

csv_file_path = input("Please enter the path: ")

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

amenity_count = {}
for row in rows:
    amenities = row['amenities']
    if amenities is not None:
        amenities = amenities.replace('{', '').replace('}', '').replace('"', '').split(',')
        for amenity in amenities:
            if amenity != '':
                if amenity in amenity_count:
                    amenity_count[amenity] += 1
                else:
                    amenity_count[amenity] = 1

sorted_amenity_count = {k: v for k, v in sorted(amenity_count.items(), key=lambda item: item[1], reverse=True)}

print("Top 10 most popular amenities or features provided by Airbnb hosts:")
for i, (amenity, count) in enumerate(sorted_amenity_count.items()):
    if i == 10:
        break
    print(f"{i+1}. {amenity}: {count}")
```

```python
#C
import csv
from sklearn.preprocessing import LabelEncoder

csv_file_path = input("Please enter the path: ")

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

le = LabelEncoder()
le.fit_transform
location_price = {}
location_count = {}
for row in rows:
    location = row['host_location']
    price = row['price']
    if location in location_price:
        location_price[location] += float(price)
        location_count[location] += 1
    else:
        location_price[location] = float(price)
        location_count[location] = 1

location_avg_price = {}
for location, price in location_price.items():
    count = location_count[location]
    avg_price = price / count
    location_avg_price[location] = avg_price

sorted_location_avg_price = {k: v for k, v in sorted(location_avg_price.items(), key=lambda item: item[1], reverse=True)}

print("Average price of stay for each location:")
for i, (location, avg_price) in enumerate(sorted_location_avg_price.items()):
    print(f"{i+1}. {location}: {avg_price:.2f}")
```

```python
#D
import csv

csv_file_path = input("Please enter the path: ")

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

locations = set([row['host_location'] for row in rows])
for location in locations:
    total_rating = 0
    count = 0
    for row in rows:
        if row['host_location'] == location:
            total_rating += float(row['review_scores_rating'])
            count += 1
    if count > 0:
        average_rating = total_rating / count
        print(f"{location}: {average_rating:.2f}")
    else:
        print(f"No reviews for {location}")
```

```python
import csv

csv_file_path = input("Please enter the path: ")

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

amenities = {}
for row in rows:
    if row['amenities']:
        row_amenities = row['amenities'].replace('{', '').replace('}', '').replace('"', '').split(',')
        for amenity in row_amenities:
            if amenity.strip() not in amenities:
                amenities[amenity.strip()] = 1
            else:
                amenities[amenity.strip()] += 1
top_10_amenities = sorted(amenities.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top 10 amenities or features:")
for amenity, count in top_10_amenities:
    print(f"{amenity}: {count}")

property_types = {}
for row in rows:
    if row['property_type']:
        property_type = row['property_type']
        if property_type not in property_types:
            property_types[property_type] = [float(row['price'].replace('$', '').replace(',', ''))]
        else:
            property_types[property_type].append(float(row['price'].replace('$', '').replace(',', '')))
print("Average price of stay:")
for property_type, prices in property_types.items():
    average_price = sum(prices) / len(prices)
    print(f"{property_type}: ${average_price:.2f}")

# Analyze the number of reviews for each host
hosts = {}
for row in rows:
    if row['host_id']:
        host_id = row['host_id']
        if host_id not in hosts:
            hosts[host_id] = [int(row['number_of_reviews'])]
        else:
            hosts[host_id].append(int(row['number_of_reviews']))
print("Number of reviews for each host:")
for host_id, reviews in hosts.items():
    total_reviews = sum(reviews)
    print(f"Host {host_id}: {total_reviews} reviews")
```

```python
#Q-3-A
import csv
import matplotlib.pyplot as plt

csv_file_path = input("Please enter the path:")

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

bedroom_counts = {}
for row in rows:
    if row['bedrooms']:
        bedrooms = int(row['bedrooms'])
        if bedrooms not in bedroom_counts:
            bedroom_counts[bedrooms] = 1
        else:
            bedroom_counts[bedrooms] += 1

labels = bedroom_counts.keys()
sizes = bedroom_counts.values()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Proportion Number of Bedrooms")
plt.axis('equal')
plt.show()
```

```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

csv_file_path = input("Please enter the path: ")

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

df = pd.DataFrame(rows)

df['last_review'] = pd.to_datetime(df['last_review'])

df['year_month'] = df['last_review'].dt.to_period('M')

df = df.groupby(['year_month', 'room_type'])['price'].mean().reset_index()

df = df.pivot(index='year_month', columns='room_type', values='price')

plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, fmt='.0f', cmap='coolwarm')
plt.title("Average Prices by Room Type and Month")
plt.xlabel("Room Type")
plt.ylabel("Month-Year")
plt.show()
```
